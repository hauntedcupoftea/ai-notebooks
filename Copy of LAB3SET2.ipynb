```python
# import and initialize everything needed.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
```

image.png

```python
Xi = np.array([8, 3, 2, 10, 11, 3, 6, 5])
Yi = np.array([4, 12, 1, 12, 9, 4, 6, 1])

sum_x = np.sum(Xi)
sum_y = np.sum(Yi)
sum_xy = np.sum(np.multiply(Xi,Yi))
sum_xx = np.sum(np.multiply(Xi,Xi))

number_of_records = len(Xi)
c = (sum_y*sum_xx - sum_x*sum_xy) / (number_of_records*sum_xx - sum_x*sum_x)
m = (number_of_records*sum_xy - sum_x*sum_y) / (number_of_records*sum_xx - sum_x*sum_x)

plt.figure(figsize = (10,8))
plt.plot(Xi, Yi, 'b.')
plt.plot(Xi, m*Xi + c, 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
h = np.array([2, 3, 5, 7, 9])
ice = np.array([4, 5, 7, 10, 15])

m, c = np.polyfit(h, ice, 1)

plt.figure(figsize = (10,8))
plt.plot(h, ice, 'b.')
plt.plot(h, m*h + c, 'r')
plt.xlabel('hours of sunshine')
plt.ylabel('ice cream sold')
plt.show()
print("number of ice-cream sold on an 8 hour day would be " + str(round(m*8+c)))
```

```python
exp = np.array([16, 12, 18, 4, 3, 10, 5, 12])
perf = np.array([8, 8, 9, 6, 6, 8, 7, 8.3])
index = np.where(perf > 10)
exp = np.delete(exp, index)
perf = np.delete(perf, index)
m, c = np.polyfit(exp, perf, 1)

plt.figure(figsize = (10,8))
plt.plot(exp, perf, 'b.')
plt.plot(exp, m*exp + c, 'r')
plt.xlabel('experience')
plt.ylabel('performance')
plt.show()
print("Slope of best fit line is: " + str(m) + "\n")
print("Intercept of best fit line is: " + str(c) + "\n")
print("Estimation of performance for a faculty with 25 years of experience: " + str(round(m*25+c)))
```

```python
from sklearn import linear_model as lm 
reg = lm.LinearRegression()

X = np.array([10, 12, 14, 16, 18, 20, 22, 24, 25, 28]).reshape(-1, 1)
Y = np.array([1, 5, 10, 15, 20, 25, 30, 35, 40, 45])

model = reg.fit(X, Y)

print('intercept:', model.intercept_)
print('slope:', model.coef_[0]) 
print("y = " + str(model.coef_[0]) + "*x" + str(model.intercept_))

plt.figure(figsize = (10,8))
plt.plot(X, Y, 'b.')
plt.plot(X, model.coef_[0]*X + model.intercept_, 'r')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

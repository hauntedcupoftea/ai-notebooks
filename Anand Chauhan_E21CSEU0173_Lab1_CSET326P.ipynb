This notebook implements fuzzy set operations in the standard python library using Dictionaries as the base data structure.

```python
while True:
    # function definitions go here
    def single_input():
        X = dict()
        m = int(input("How many elements in the set?: "))
        for _ in range(m):
            key = input("Enter element: ")
            value = float(input(f"Enter membership value of {key}: "))
            if value < 0 or value > 1:
                raise ValueError("Membership value cannot be outside range [0, 1].")
            X[key] = value
        return X

    def dual_input():
        X = dict()
        Y = dict()
        m = int(input("How many elements in BOTH sets each?: "))
        for _ in range(m):
            key = input("Enter element: ")
            value = float(input(f"Enter membership value of {key}: "))
            if value < 0 or value > 1:
                raise ValueError("Membership value cannot be outside range [0, 1].")
            X[key] = value
        print("\n...set one complete...\n")
        for _ in range(m):
            key = input("Enter element: ")
            value = float(input(f"Enter membership value of {key}: "))
            if value < 0 or value > 1:
                raise ValueError("Membership value cannot be outside range [0, 1].")
            Y[key] = value
        return (X, Y)

    def comp():
        X = single_input()
        print(f"\nThe set is: {X}\n")
        comp = X
        for key in X:
            comp[key] = 1 - X[key]
        print(f"=> Complement of given set: {comp}")

    def scale():
        scale = float(input("Enter the scaling factor (between 0 and 1): "))
        if scale < 0 or scale > 1:
            print("Scalar cannot be outside range [0, 1].")
            return
        X = single_input()
        print(f"\nThe set is: {X}\n")
        comp = X
        for key in X:
            comp[key] = scale*X[key]
        print(f"Scalar Product of given set: {comp}")

    def union():
        X, Y = dual_input()
        print(f"\nThe first set is: {X}\nThe second set is: {Y}\n")
        union = dict()
        for Xkey, Ykey in zip(X, Y):
            Xval = X[Xkey]
            Yval = Y[Ykey]
            if Xval > Yval:
                union[Xkey] = Xval
            else:
                union[Ykey] = Yval
        print(f"Union of Sets is: {union}")

    def inter():
        X, Y = dual_input()
        print(f"\nThe first set is: {X}\nThe second set is: {Y}\n")
        inter = dict()
        for Xkey, Ykey in zip(X, Y):
            Xval = X[Xkey]
            Yval = Y[Ykey]
            if Xval < Yval:
                inter[Xkey] = Xval
            else:
                inter[Ykey] = Yval
        print(f"Intersection of Sets is: {inter}")

    def dot():
        X, Y = dual_input()
        print(f"\nThe first set is: {X}\nThe second set is: {Y}\n")
        dot = dict()
        for elem in X:
            if elem in Y:
                dot[elem] = X[elem]*Y[elem]
        print(f"Dot Product of Sets is: {dot}")

    def setsum():
        X, Y = dual_input()
        print(f"\nThe first set is: {X}\nThe second set is: {Y}\n")
        dot = dict()
        for elem in X:
            if elem in Y:
                dot[elem] = X[elem] + Y[elem] - X[elem]*Y[elem]
        print(f"Sum of Sets is: {dot}")

    def diff():
        X, Y = dual_input()
        print(f"\nThe first set is: {X}\nThe second set is: {Y}\n")
        comp = dict()
        result = dict()
        for key in X:
            comp[key] = 1 - X[key]
        for Xkey, Ykey in zip(comp, Y):
            Xval = comp[Xkey]
            Yval = Y[Ykey]
            if Xval < Yval:
                result[Xkey] = Xval
            else:
                result[Ykey] = Yval
        print(f"Set Difference of Sets is: {result}")

    # master dictionary to call everything
    master = {1 : comp, 2 : scale, 3 : union, 4 : inter,
              5 : dot, 6 : setsum, 7 : diff}

    # string to take input
    print("""\n
Select a Fuzzy Set Operation (or input -1 to terminate):
Operations on One Fuzzy Set -
    1 - Complement of One Fuzzy Set
    2 - Scalar Product on One Fuzzy Set
Operations on Two Fuzzy Sets -
    3 - Union of Two Fuzzy Sets
    4 - Intersection of Two Fuzzy Sets
    5 - Vector Product of Two Fuzzy Sets
    6 - Sum of Two Fuzzy Sets
    7 - Difference of Two Fuzzy Sets
8 or -1: Terminate program
    \n""")
    choice = int(input("Enter the integer of your choice: "))
    if choice == -1 or choice == 8:
        print("Exiting...")
        break
    elif choice > 0 and choice < 8:
        master[choice]()
    else:
        print("Invalid input!")
        break
```

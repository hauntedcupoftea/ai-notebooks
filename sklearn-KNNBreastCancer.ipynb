```python
# import libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import scale

# read dataset
dataset = pd.read_csv('BreastCancerData.csv')

# print head of dataset
dataset.head()
```

```python
# Extract independent and dependent variables from the dataset
X = dataset.iloc[:, 2:-1].values
y = dataset.iloc[:, 1].values

# Scale the values
X = scale(X)

# checking if the number of records is still the same
print(np.shape(X)[0] - np.shape(y)[0])  
```

```python
# Split data into train and test groups
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 15/569, random_state = 42)
```

```python
# Implement KNN Classifier
model = KNeighborsClassifier()

# Fit model to training data
model.fit(X_train, y_train)
```

```python
# Predictions for KNN Classifier using test data
y_predict = model.predict(X_test)
```

```python
# Making the confusion matrix
cf = confusion_matrix(y_predict, y_test)
```

```python
# Printing the Accuracy Score
accuracy_score(y_predict, y_test)
```

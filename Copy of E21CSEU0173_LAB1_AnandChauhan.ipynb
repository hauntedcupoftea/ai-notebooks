1. Write a program to convert any decimal number into Binary.

  Sample input: 13 base 10
  
  Sample output: 1101 base 2

```python
# Anand Chauhan, E21CSEU0173

x = int(input())
y = ""
z= ""
while (x > 0):
  y += str(x % 2)
  x = x//2

for i in range(len(y)-1, -1, -1):
  z += y[i]

print(z)
```

2. Write a program to convert any binary number into a decimal number

    Sample input: 1100 base 2

    Sample output: 12 base 10

```python
# Anand Chauhan, E21CSEU0173

x = int(input())
y = 0
z = 0
n = 0

while x > 0:
  y = x % 2
  z += y*(2**n)
  n += 1
  x = x // 10

print(z)
```

3. Write a program to convert any binary number to octal number.

    Sample Input: 1100 base 2

    Sample Output: 14 base 8

```python
# Anand Chauhan, E21CSEU0173

x = int(input())
y = 0
z = 0
n = 0

while x > 0:
  y = x % 2
  z += y*(2**n)
  n += 1
  x = x // 10

x = ""
y = ""

while z > 0:
  y += str(z % 8)
  z = z // 8

for i in range(len(y)-1, -1, -1):
  x += y[i]

print(x)
```

4. Write a program to convert any binary number to a hexadecimal number.

    Sample input: 10010 base 2
    
    Sample output: 12 base 16

```python
# Anand Chauhan, E21CSEU0173

x = int(input())
y = 0
z = 0
n = 0

while x > 0:
  y = x % 2
  z += y*(2**n)
  n += 1
  x = x // 10

x = z
y = ""
z = ""

while x > 0:
  n = x % 16
  x = x // 16
  if (n < 10):
    y += str(n)
  elif (n == 10):
    y += "A"
  elif (n == 11):
    y += "B"
  elif (n == 12):
    y += "C"
  elif (n == 13):
    y += "D"
  elif (n == 14):
    y += "E"
  else:
    y += "F"

for i in range(len(y)-1, -1, -1):
  z += y[i]

print(z)
```

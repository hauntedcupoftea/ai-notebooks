Q2. Write a program to create a 1D Array

```python
poggers = np.array([x for x in range(1, 11, 1)])
poggers
```

Q4. Write a program to find out odd numbers from the 1D array

```python
pog = poggers[poggers % 2 == 1]
pog
```

Q5. Write program to replace elements matching a condition in an array

```python
poggers[poggers % 2 == 1] = -1
poggers
```

Q6. Write a program to reshape an array

```python
poggers = poggers.reshape(2, -1)
poggers
```

Q7. Write a program to stack two arrays vertically

```python
gers = np.vstack([pog, poggers])
gers
```

Q9. Write a program to remove matching elements from an array

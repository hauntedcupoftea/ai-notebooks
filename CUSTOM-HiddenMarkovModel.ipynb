```python
import numpy as np
```

```python
startprob = np.array([0.2, 0.8])
transmat = np.array([[0.5, 0.5], [0.4, 0.6]])
emismat = np.array([[0.5, 0.4, 0.1], [0.2, 0.4, 0.4]])
```

```python
def fwd_alg(y, A=transmat, B=emismat, Pi=startprob):
    # initialize
    K = A.shape[0]
    T = len(y)
    T1 = np.empty((K, T), 'd')
    T2 = np.empty((K, T), 'B')
    T1[:, 0] = Pi * B[:, y[0]]
    T2[:, 0] = 0

    # run the dynamic program
    for i in range(1, T):
        T1[:, i] = np.max(T1[:, i - 1] * A.T * B[np.newaxis, :, y[i]].T
                          + T2[:, i - 1], 1)
        T2[:, i] = np.argmax(T1[:, i - 1] * A.T, 1)

    # extract output in reverse
    x = np.empty(T, 'B')
    x[-1] = np.argmax(T1[:, T - 1])
    for i in reversed(range(1, T)):
        x[i - 1] = T2[x[i], i]

    output = {0: 'COLD', 1: 'HOT'}
    answer = [output[i] for i in x]
    return answer, T1
```

```python
def run():
    seq = np.array(list(map(int, input("Enter Sequence (Space Seperated): ").split())))
    # convert to index
    seq = seq - 1
    answer, probas = fwd_alg(seq)
    # output format
    sep = ' -> '
    print(f"The optimal sequence of states is: {sep.join(list(answer))}")
    print(f"with state probabilites-\n\tCOLD: {probas[0, :]}\n\tHOT: {probas[1, :]}")
```

```python
# DEMO CELL
run()
```
